```python
import sklearn #nome do pacote para usar
```

```python
#Importando o dataset iris do Scikit-learn
from sklearn.datasets import load_iris #dataset utilizado com frequencia por iniciaintes
data = load_iris()
```

```python
#E ento carregamos o dataset iria como dataframe do Pandas
import pandas as pd
df = pd.DataFrame(data['data'], columns = data['feature_names']) #carregando os atributos
df['species'] = data['target'] #carregando a classe ou varivel alvo
df.head()
```

```python
df.plot()
```

```python
#Criao de um scatter plot (grfico de disperso) com duas variveis
df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)')
```

```python
# E o mesmo grfico mais complexo, como um grfico de rea., pode ser criado:
columns = ['sepal length (cm)', 'petal width (cm)', 'petal length (cm)', 'sepal width (cm)']
df[columns].plot.area()
```
